```python
alpha = 0.35
beta = 1.00
df['SES'] = 0.00
df['DES'] = 0.00
df['SES'][0] = df['Y_true'][0]
df['DES'][0] = df['Y_true'][1] - df['Y_true'][0]
df.head()
```

```python
x = 1
while x < 39:
    x += 1
    for i in range(1,39):
        df['SES'][i] = alpha * df['Y_true'][i] + (1-alpha) * (df['SES'][i-1] + df['DES'][i-1])
        df['DES'][i] = beta * ( df['SES'][i] - df['SES'][i-1] ) + (1-beta) * df['DES'][i-1]
df.head()  
```

```python
df.tail(8)  
```

```python
df['Y_pred'] = 0.00
for i in range(2,40):
    df['Y_pred'][i] = df['SES'][i-1] + df['DES'][i-1]
df.head()  
```

```python
df.tail(8) 
```

```python
df['Absolute Error'] = 0.00
for i in range(2,39):
    df['Absolute Error'][i] = abs( df['Y_pred'][i] - df['Y_true'][i] )
df.head()                                
```

```python
df['Squared Error'] = 0.00
for i in range(2,39):
    df['Squared Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Precent Error1'] = 0.00
for i in range(2,39):
    df['Precent Error1'][i] = abs ((  df['Y_true'][i] - df['Y_pred'][i] ) / df['Y_true'][i])
df.head()                                
```

```python
df.tail()
```

```python
df['Precent Error2'] = 0.00
for i in range(2,39):
    df['Precent Error2'][i] = ( (  df['Y_pred'][i] - df['Y_true'][i] ) / df['Y_true'][i-1] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Precent Error3'] = 0.00
for i in range(2,39):
    df['Precent Error3'][i] = ( (  df['Y_true'][i] - df['Y_true'][i-1] ) / df['Y_true'][i-1] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Error'] = 0.00
for i in range(2,39):
    df['Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )
df.head()                                
```

```python
df.tail() 
```

```python
df['Squared Error Diff'] = 0.00
for i in range(3,39):
    df['Squared Error Diff'][i] = ( df['Error'][i] - df['Error'][i-1] )**2
df.head()                                
```
